```python
import pandas as pd
import numpy as np

exam_data = {'name': ['Anastasia', 'Dima', 'Katherine', 'James', 'Emily', 'Michael', 'Matthew', 'Laura', 'Kevin', 'Jonas', "Elias", 'Myriam'],

'score': [12.5, 9, 16.5, np.nan, 9, 20, 14.5, np.nan , 8, 19, 20, 16],

'attempts': [1, 3, 2, 3, 2, 3, 1, 1, np.nan ,1, 3, 2],

'qualify': ['yes', 'no', 'yes', 'no', 'no', 'yes', 'yes', 'no',np.nan , 'yes', 'yes', 'yes']}



examdata = pd.DataFrame(exam_data, index  = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j', 'k', 'l'])
examdata
```

```python
examdata.isnull().sum()
```

```python
examdata.info()
```

```python
examdata['score'].fillna(examdata['score'].mode(), inplace = True)
```

```python
examdata['attempts'].fillna(examdata['attempts'].mode(), inplace = True)

#Changes were not effected using this method
```

```python
examdata.loc['d', 'score'] = 12.0
```

```python
examdata.loc['h', 'score'] = 13.5
```

```python
examdata.loc['i', 'attempts'] = 2.0
```

```python
examdata.loc['i', 'qualify'] = 'no'
```

```python
examdata

#To show that changes have been effected
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
coder = LabelEncoder()

examdata['qualify'] = coder.fit_transform(examdata['qualify'])
examdata
```
