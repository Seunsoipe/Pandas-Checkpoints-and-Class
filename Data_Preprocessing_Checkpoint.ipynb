```python
tested = pd.read_csv(io.BytesIO(uploaded['tested.csv']))

tested.drop(['Cabin'], axis=1)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
coder = LabelEncoder()

tested['Sex'] = coder.fit_transform(tested['Sex'])

tested.head()
```

```python
embarked_dict = {
    'Q': 1,
    'S': 2,
    'C': 3
}

tested['Embarked'].replace(embarked_dict, inplace = True)
tested
```
